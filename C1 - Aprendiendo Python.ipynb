**Veamos las funciones con ejercicios**

```python
# Vamos a crear un modelo muy simple (adems de errneo) que nos va a decir la evolucin de un virus 
# La idea esta inspirada en este vdeo https://www.youtube.com/watch?v=Kas0tIxDvrg

# Queremos crear una funcin que, dado un determinado nmero de das 
# nos devuelva una lista con la simulacin con del nmero de infectados por da:

def simulacion_infectados(num_dias, prob_infecc):
    
    """
    num_dias: dias totales de simulacin
    prob_infecc : probabilidad de que dada una aproximacin entre personas acabe en infeccin
    """

    num_casos = [1] # Inicializamos la lista de casos con un nico caso el primer da
    
    # Por cada da que pasa calculamos el nmero de nuevos infectados
    for _ in range(num_dias):
        
        num_casos_total_ayer = num_casos[-1] # Ultimo valor de la lista
        num_casos_nuevos_hoy = num_casos_total_ayer * prob_infecc # Incremento segn probabilidad de infeccin
        num_casos_nuevos_hoy = round(num_casos_nuevos_hoy) # Redondeamos a un valor absoluto (no puede haber 1.4 infectados) 
        num_casos_total_hoy = num_casos_total_ayer + num_casos_nuevos_hoy # Sumamos al total los casos nuevos
        num_casos.append(num_casos_total_hoy) # A la lista de nmero de casos aadimos los casos de hoy
        
    # Cuando hemos terminado de simular todos los das devolvemos los resultados
    return num_casos
```

```python
PROBABILIDAD_DE_INFECCION = 0.6
```

```python
simulacion_infectados(0, PROBABILIDAD_DE_INFECCION)
```

```python
simulacion_infectados(1, PROBABILIDAD_DE_INFECCION)
```

```python
simulacion_infectados(10, PROBABILIDAD_DE_INFECCION)
```

```python
simulacion_infectados(20, PROBABILIDAD_DE_INFECCION)
```

```python
simulacion_infectados(40, PROBABILIDAD_DE_INFECCION)
```

### Clases

Las clases son objetos (puedes imaginarlo como una caja) que contienen una serie de variables y funciones.
En python, para definir una clase, se coloca *class* delante del nombre de la clase.

```python
class Persona():
    
    # La funcin __init__ est predeterminada en python y se ejecuta una vez cuando creamos la clase.
    # Todas las funciones deben incluir la clase a la que pertenecen dentro de sus parametros: self.
    
    def __init__(self):
        
        # Las variables dentro de una clase se indican con el prefijo self
        self.nombre = "Charlie"
        self.apellido = "Brown"
        
    def presentarse(self):
        
        # Cuando ejecutemos la siguiente funcin la clase Persona() imprimir un saludo
        print("Hola!, me llamo", self.nombre, self.apellido)
```

```python
# creamos una variable de tipo Persona
persona = Persona()
```

```python
# ejecutamos la funcin de presentarse
persona.presentarse()
```

```python
class Persona():
    
    # La funcin __init__ est predeterminada en python y se ejecuta una vez cuando creamos la clase.
    # Todas las funciones deben incluir la clase a la que pertenecen dentro de sus parametros: self.
    
    def __init__(self, nombre, apellido):
        
        # Las variables dentro de una clase se indican con el prefijo self
        self.nombre = nombre
        self.apellido = apellido
        
    def presentarse(self):
        
        # Cuando ejecutemos la siguiente funcin la clase Persona() imprimir un saludo
        print("Hola!, me llamo", self.nombre, self.apellido)
        
    def presentarse_a_otra_persona(self, persona):
        
        # Cuando ejecutemos la siguiente funcin la clase Persona() imprimir un saludo personalizado
        print("Hola", persona.nombre, "!, me llamo", self.nombre, self.apellido)
```

```python
# Creamos la clase persona pasndole los argumentos de inicio
persona1 = Persona(nombre = "Pepito", apellido = "Grillo")
persona1.presentarse()
```

```python
persona2 = Persona(nombre = "Pepa", apellido = "Garca")
persona2.presentarse_a_otra_persona(persona1)
```

```python
persona1.presentarse_a_otra_persona(persona2)
```
